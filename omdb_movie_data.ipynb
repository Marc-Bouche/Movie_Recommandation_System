```python
# Column renaming
df1 = df1.rename(columns={'Duration (min)': 'Runtime',
                         'Review Count': 'Review_count',
                         'Review Title':'Review_title','Rating':'imdb_rating'})
df1['Year'] = df1['Year'].astype(str).str.replace('.0', '', regex=False)

cols = df1.columns.tolist()
cols.insert(len(cols) - 1, cols.pop(cols.index('Cast')))
df1 = df1[cols]


columns = df1.columns.tolist()
columns.remove('Poster')
columns.append('Poster')
df1 = df1[columns]

# Create new actor columns
df1['Actor1'] = [actors.split(',')[0] if isinstance(actors, str) and len(actors.split(',')) > 0 else np.nan for actors in df1['Cast']]
df1['Actor2'] = [actors.split(',')[1] if isinstance(actors, str) and len(actors.split(',')) > 1 else np.nan for actors in df1['Cast']]
df1['Actor3'] = [actors.split(',')[2] if isinstance(actors, str) and len(actors.split(',')) > 2 else np.nan for actors in df1['Cast']]
df1['Actor4'] = [', '.join(actors.split(',')[3:]) if isinstance(actors, str) and len(actors.split(',')) > 3 else np.nan for actors in df1['Cast']]
```
